```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
historyfiles = glob.glob('*.csv')
historyfiles
```

```python
for historyfile in historyfiles:
    history = pd.read_csv(historyfile)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize = [10,12])
    ax1.plot(history.loss, label = 'train_loss')
    ax1.plot(history.val_loss, label = 'val_loss')
    ax1.grid()
    ax1.legend(loc = 1, bbox_to_anchor=(1.16, .7))
    ax1.set_title('loss', fontsize=18)
    
    ax2.plot(history.acc, label = 'train_acc')
    ax2.plot(history.val_acc, label = 'val_acc')
    ax2.grid()
    ax2.legend(loc = 1, bbox_to_anchor=(1.16, .7))
    ax2.set_title('accuracy', fontsize=18)
    
    print(historyfile)
    plt.show()
    
```
